```python
num = [10, 7, 13, 4, 19, 20, 23]

print("Prime Numbers are :")

for n in num:

    if n > 1:

        prime = True

        for i in range(2, n):

            if n % i == 0:
                prime = False
                break

        if prime:
            print(n)
```

```python
def check_numbers(num):

    even_count = 0
    odd_sum = 0

    for i in num:

        if i % 2 == 0:
            even_count = even_count + 1

        else:
            odd_sum = odd_sum + i

    print("Count of Even Numbers :", even_count)
    print("Sum of Odd Numbers :", odd_sum)


numbers = [10, 15, 20, 25, 30, 35]

check_numbers(numbers)
```

```python
def simple_interest(principal, rate=5, time=2):

    si = (principal * rate * time) / 100

    print("Simple Interest :", si)


simple_interest(1000)

simple_interest(principal=2000, rate=10, time=3)
```

```python
class Student:

    def __init__(self, name, marks):

        self.name = name
        self.marks = marks

    def calculate_grade(self):

        if self.marks >= 90:
            print("Grade : A")

        elif self.marks >= 75:
            print("Grade : B")

        elif self.marks >= 50:
            print("Grade : C")

        else:
            print("Grade : D")


s1 = Student("Ajit", 85)

print("Name :", s1.name)
print("Marks :", s1.marks)

s1.calculate_grade()
```

```python
class Student:

    def __init__(self, name, marks):

        self.name = name
        self.marks = marks

    def calculate_grade(self):

        if self.marks >= 90:
            print("Grade : A")

        elif self.marks >= 75:
            print("Grade : B")

        elif self.marks >= 50:
            print("Grade : C")

        else:
            print("Grade : D")


s1 = Student("Ajit", 85)
s2 = Student("Rohit", 92)

print("Student 1")
print("Name :", s1.name)
print("Marks :", s1.marks)
s1.calculate_grade()

print()

print("Student 2")
print("Name :", s2.name)
print("Marks :", s2.marks)
s2.calculate_grade()
```

```python
class BankAccount:

    def __init__(self, balance):

        self.__balance = balance

    def deposit(self, amount):

        self.__balance = self.__balance + amount

        print("Amount Deposited :", amount)

    def withdraw(self, amount):

        if amount <= self.__balance:

            self.__balance = self.__balance - amount

            print("Amount Withdrawn :", amount)

        else:

            print("Insufficient Balance")

    def show_balance(self):

        print("Current Balance :", self.__balance)


account = BankAccount(5000)

account.deposit(2000)

account.withdraw(1000)

account.show_balance()
```

```python
try:

    file = open("10,20,30,40,50", "r")

    total = 0
    count = 0

    for line in file:

        num = int(line)

        total = total + num

        count = count + 1

    average = total / count

    print("Total :", total)
    print("Average :", average)

    file.close()

except FileNotFoundError:

    print("Error : File Does Not Exist")
```
