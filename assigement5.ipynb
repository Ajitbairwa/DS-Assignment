1. Explain pandas and its role in Data Analysis and Machine Learning.


Pandas is an open-source Python library used for data analysis and data manipulation. It provides powerful data structures like Series (1-dimensional) and DataFrame (2-dimensional) to store and work with data easily.


Role of Pandas in Data Analysis:
  It is used to read data from files such as CSV, Excel, and SQL.
  It helps in cleaning data by handling missing values and duplicate records.
  It allows filtering, sorting, and selecting specific rows and columns.
  It is used to calculate statistical values such as mean, median, maximum, and minimum.
  It makes data analysis faster and easier.
  
 Role of Pandas in Machine Learning:
  Pandas is used to prepare data before training a machine learning model.
  It helps in selecting input features and target columns.
  It converts raw data into a proper format for machine learning algorithms.
  It works well with libraries like NumPy, Matplotlib, Seaborn, and Scikit-learn.
  It is commonly used for data preprocessing before model training.

2. Differentiate between:
● Series and DataFrame
● loc[] and iloc[] functions in Pandas

(a) Difference between Series and DataFrame

Series:
A Series is a one-dimensional data structure in Pandas. It contains only one column of data and is created using pd.Series(). It is useful for storing a single list of values.


DataFrame:
A DataFrame is a two-dimensional data structure in Pandas. It contains multiple rows and columns and is created using pd.DataFrame(). It is used to store and manage tabular data, similar to an Excel sheet.


Difference:

Series is one-dimensional, while DataFrame is two-dimensional.
Series contains a single column, whereas DataFrame contains multiple columns.
Series is suitable for a single type of data, while DataFrame can store different types of data in different columns.


(b) Difference between loc[] and iloc[]


loc[]:
The loc[] function is used to access data using row and column labels (names). It works with the actual labels present in the DataFrame.


iloc[]:
The iloc[] function is used to access data using row and column index positions. It works only with integer index values.


Difference:


loc[] uses labels or names to select data.
iloc[] uses integer index positions to select data.
loc[] is label-based indexing, whereas iloc[] is integer-based indexing.

Create a Pandas DataFrame using the dataset below: Name Age Marks Amit 20 85 Neha 21 90 Rahul 19 78 Sneha 22 88 a) Tasks: ● Display first 2 rows ● Display only Name and Marks columns ● Add a new column named “Result” ● Display students with marks greater than 80 b) Expected Concepts: ● pd.DataFrame() ● head() ● Column Selection ● Conditional Filtering

```python
import pandas as pd

data = {
    "Name": ["Amit", "Neha", "Rahul", "Sneha"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

print("\nFirst 2 Rows")
print(df.head(2))

print("\nName and Marks Columns")
print(df[["Name", "Marks"]])

df["Result"] = "Pass"

print("\nDataFrame after adding Result column")
print(df)

print("\nStudents with Marks greater than 80")
print(df[df["Marks"] > 80])
```

4. Create a DataFrame using the dataset below:
Name Age Salary
A 25 30000
B NaN 40000
C 27 NaN
D 24 35000
a) Tasks:
● Identify missing values
● Fill missing Age with average age
● Fill missing Salary with median salary
● Display cleaned dataset
b) Expected Concepts:
● isnull()
● fillna()
● mean()
● median()

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [25, np.nan, 27, 24],
    "Salary": [30000, 40000, np.nan, 35000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

print("\nMissing Values")
print(df.isnull().sum())

avg_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(avg_age)

median_salary = df["Salary"].median()
df["Salary"] = df["Salary"].fillna(median_salary)

print("\nCleaned DataFrame")
print(df)
```
