1. Find the largest, smallest, and second largest number from 10 user 
inputs.  

```python
num = []

for i in range(10):
    n = int(input("Enter Number : "))
    num.append(n)

print("Numbers are :", num)

num.sort()

print("Smallest Number :", num[0])

print("Second Largest Number :", num[-2])

print("Largest Number :", num[-1])
```

2. Count vowels, consonants, digits, and special characters in a string.

```python
string = input("Enter a String : ")

vowels = 0
consonants = 0
digits = 0
special = 0

for ch in string:

    if ch.lower() in "aeiou":
        vowels += 1

    elif ch.isalpha():
        consonants += 1

    elif ch.isdigit():
        digits += 1

    else:
        special += 1

print("Vowels :", vowels)
print("Consonants :", consonants)
print("Digits :", digits)
print("Special Characters :", special)
```

3. Check whether a number is an Armstrong number.  

```python
num = int(input("Enter a Number : "))

temp = num
sum = 0

while temp > 0:

    digit = temp % 10
    sum = sum + (digit ** 3)
    temp = temp // 10

if num == sum:
    print("Armstrong Number")

else:
    print("Not an Armstrong Number")
```

4. Find duplicate elements and their frequency in a tuple.  

```python
tup = (10, 20, 30, 20, 40, 10, 50, 20)

print("Tuple is :", tup)

for i in tup:

    if tup.count(i) > 1:
        print("Element :", i, "Frequency :", tup.count(i))
```

5. Find the topper from a tuple of student names and marks.

```python
students = (("Ajit", 85),
            ("Rohit", 92),
            ("Mohit", 78),
            ("Aman", 95),
            ("Karan", 88))

topper_name = ""
topper_marks = 0

for i in students:

    if i[1] > topper_marks:

        topper_marks = i[1]
        topper_name = i[0]

print("Topper Name :", topper_name)
print("Marks :", topper_marks)
```

6. Count the frequency of each word in a sentence using a dictionary.  

```python
sentence = input("Enter a Sentence : ")

words = sentence.split()

freq = {}

for i in words:

    if i in freq:
        freq[i] = freq[i] + 1

    else:
        freq[i] = 1

print("Word Frequency :")

for key, value in freq.items():
    print(key, ":", value)
```

7. Display students scoring above the average marks from a dictionary.

```python
students = {
    "Ajit": 85,
    "Rohit": 70,
    "Mohit": 90,
    "Aman": 60,
    "Karan": 95
}

total = 0

for marks in students.values():
    total = total + marks

average = total / len(students)

print("Average Marks :", average)

print("Students Scoring Above Average :")

for name, marks in students.items():

    if marks > average:
        print(name, ":", marks)
```

8. Create a menu-driven product management system using a dictionary. 

```python
products = {}

while True:

    print("\n1. Add Product")
    print("2. View Products")
    print("3. Update Product")
    print("4. Delete Product")
    print("5. Exit")

    choice = int(input("Enter Your Choice : "))

    if choice == 1:

        name = input("Enter Product Name : ")
        price = int(input("Enter Product Price : "))

        products[name] = price

        print("Product Added Successfully")

    elif choice == 2:

        print("Products List :")

        for name, price in products.items():
            print(name, ":", price)

    elif choice == 3:

        name = input("Enter Product Name : ")

        if name in products:

            price = int(input("Enter New Price : "))
            products[name] = price

            print("Product Updated Successfully")

        else:
            print("Product Not Found")

    elif choice == 4:

        name = input("Enter Product Name : ")

        if name in products:

            del products[name]

            print("Product Deleted Successfully")

        else:
            print("Product Not Found")

    elif choice == 5:

        print("Program Ended")
        break

    else:
        print("Invalid Choice")
```

9. Check whether a triangle is Equilateral, Isosceles, or Scalene. 

```python
side1 = int(input("Enter First Side : "))
side2 = int(input("Enter Second Side : "))
side3 = int(input("Enter Third Side : "))

if side1 == side2 and side2 == side3:

    print("Equilateral Triangle")

elif side1 == side2 or side2 == side3 or side1 == side3:

    print("Isosceles Triangle")

else:

    print("Scalene Triangle")
```
